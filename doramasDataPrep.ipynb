```python
import json

# Dados da lista (suponho que você já tenha definido a lista_informacoes)

# Função para extrair informações de cada item
def extrair_informacoes(item):
    item_info = {}
    lines = item.strip().split('\n\n')
    for line in lines:
        parts = line.split(': ', 1)
        if len(parts) == 2:
            key, value = parts
            item_info[key] = value
    return item_info

# Converter para JSON
lista_json = [extrair_informacoes(item) for item in lista_informacoes]

json_data = json.dumps(lista_json, indent=4, ensure_ascii=False)

# Escrever o JSON em um arquivo
with open('resultado_de_conversao_doramas.json', 'w', encoding='utf-8') as f:
    f.write(json_data)
```

```python
import json
import os
from langchain.schema import Document


with open(os.path.abspath('resultado_de_conversao_doramas.json'), 'r',encoding='utf-8') as json_file:
    lista_informacoes = json.load(json_file)

documents_to_index = []

for doc in lista_informacoes:
    document = Document(
        page_content=
            f"NOME DORAMA: {doc['NOME DORAMA']}\n"
            f"URL: {doc['URL']}\n"
            f"GENERO: {doc['GENERO']}\n"
            f"DURACAO: {doc['DURACAO']}\n"
            f"NUMERO DE AVALIACOES: {doc['NUMERO DE AVALIACOES']}\n"
            f"SINOPSE: {doc['SINOPSE']}\n"
            f"DATA INICIO: {doc['DATA INICIO']}\n"
            f"DATA FIM: {doc['DATA FIM']}\n"
            f"QUANTIDADE TEMPORADAS: {doc['QUANTIDADE TEMPORADAS']}\n"
            f"EPISODIOS: {doc['EPISODIOS']}\n"
            f"CRIADOR: {doc['CRIADOR']}\n"
            f"ELENCO: {doc.get('ELENCO')}\n",
        metadata={
            'NOME DORAMA': doc['NOME DORAMA'],
            'URL': doc['URL']
        }
    )
    documents_to_index.append(document)
```
